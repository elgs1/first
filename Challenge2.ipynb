```python
#import Bokeh and pandas
from bokeh.models import ColumnDataSource
from bokeh.plotting import figure
from bokeh.io import output_file, show
import pandas as pd
import datetime
```

```python
df=pd.read_csv("c1_temperature_data.csv")

df["Time"] = pd.to_datetime(df["Time"],dayfirst=True)

p=figure(width=900,height=600,x_axis_type="datetime")

p.title.text = "Temperature during time period from data set 1"
p.title.align = 'center'
p.xaxis.axis_label = "Date"
p.yaxis.axis_label = "Temperature (°C)"

data_source = ColumnDataSource(df)

p.circle(df["Time"],df["Temp"],color="DeepPink",size=2,alpha=0.5)
p.line(df["Time"],df["Temp"],color="DeepPink")

output_file("timeseries.html")
show(p)
```

```python
df.plot(x='Time', y='Temp');
```
